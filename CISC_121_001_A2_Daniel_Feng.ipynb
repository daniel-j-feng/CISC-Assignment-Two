### 1. Understanding & Breakdown

Problem Decomposition: This problem requires three main functional groups. The first is to figure out how to circle the queue, if the kid does not eat the sandwich. The second issue revolves around revolving the students and comparing it to the top of the sandwich stack. Finally, the third issue is to identify when the kids have finished their rotations, and it's time to count how many kids are left without food, without continually rotating the queue indefinitely.

Pattern Recognition: We would always need to compare the kid's preference to the top of the stack, over and over again. Additionally, if the kid does not prefer the sandwich, the action of dequeueing the kid and enqueing it again would need to be repeated across functions.

Abstraction: The most important tasks that need to be handled by the code would be the comparison of the queue preference and the top of the stack, as well as the movement of queue elements from the front to the back.

Algorithmic Thinking: The two inputs would be the array used as a stack containing sandwich types, and an array that would be used as a queue containing the student preferences. The output would be an integer denoting how many kids were not paired with a sandwich, in which the constraint would be that all inputs are either 1 or 0, and the queue array must be treated circularly. Additionally the student array and sandwich array will be of same length.
